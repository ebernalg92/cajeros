# Variables reelevantes para la colocación de cajeros

### Empezamos cargando las respectivas bases que contiene informacion acerca de los cajeros y su entorno

```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('base1.csv', sep=',',encoding='latin-1').fillna(value = 0)
df2 = pd.read_csv('base2.csv',sep=',',encoding='latin-1').fillna(value = 0)
df3 = pd.read_csv('sucursales2.csv',sep=',',encoding='latin-1').fillna(value = 0)
df4 = pd.read_csv('base3.csv',sep=',',encoding='latin-1').fillna(value = 0)
```

### Se unen las bases de datos en base a la Referencia de la Sucursal

```python
df=pd.merge(df2,df1,on="Ref")
```

```python
df=pd.merge(df,df4,on="Ref").fillna("")
```

```python
df.head()
```

### El indice de la base de datos total es el identificador del cajero

```python
dfc=df.copy()
```

```python
df['indice'] = df['ID CAJERO']
df.set_index('indice', inplace=True)
```

### El interes radica en las transacciones de otros bancos en nuestro cajero

```python
df = pd.concat([df['Otros bancos' ],df.drop('Otros bancos' ,axis=1)], axis=1)
```

```python
del df['ID CAJERO']
del df['Ref']
```

### Asi queda la base de datos final, considerando ya la correlación de las variables y su contribucion real al problema

```python
df
```

```python
df=df.fillna(0).dropna().astype(np.float32)
```

### Descripción estadistica de la base de datos

```python
df.describe()
```

### Dividimos la base de datos en un conjunto de prueba y otro de entrenamiento 

```python


from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor

X, y = df.iloc[:,1:], df.iloc[:, 0]




X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

### Se aplica un algoritmo de Random Forest para observar la reelevancia de las variables

```python
forest = RandomForestRegressor(n_estimators=3000, random_state=2)
model=forest.fit(X_train, y_train)
```

```python
print("Exactitud en el conjunto de entrenamiento: {:.3f}".format(forest.score(X_train, y_train)))
print("Exactitud en el conjunto de prueba: {:.3f}".format(forest.score(X_test, y_test)))

#print("Feature importances:\n{}".format(forest.feature_importances_))

headers = ["name", "score"]
values = sorted(zip(X_train.columns, model.feature_importances_*100), key=lambda x: x[1] * -1)
```

### Se muestra la reelevancia númerica de las variables

```python
Variables=pd.DataFrame(values,columns=["nombre", "Porcentaje "])
print(Variables)
```

### Es necesario ver cuales de estas variables hacen que las transacciones suban o bajen, por lo que se utiliza ahora una regression lineal que nos permita acercarnos a un modelo

```python
from sklearn.linear_model import Lasso
lin_reg=Lasso(alpha=0.1,max_iter=100000,normalize=True)
lin_reg.fit(X,y)
```

```python
print("Exactitud en el conjunto de entrenamiento: {:.3f}".format(lin_reg.score(X_train, y_train)))
print("Exactitud en el conjunto de prueba: {:.3f}".format(lin_reg.score(X_test, y_test)))
```
